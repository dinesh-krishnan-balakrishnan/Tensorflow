### Imports

__NOTE:__ _Keras_ is included within the _Tensorflow 2.0_ framework.

```python
import tensorflow as tf
import numpy as np
```

- - - -

### Creating a Neural Network

* **Sequential Neural Network:** : A neural network model structure that consists of individual layers defined in sequence. The first layer consists of the input shape, defined by the *input_shape* parameter passed into it.

* **Dense Layer** : A layer of fully connected neurons.

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 1, input_shape=[1])
])

# The metrics attribute specifies what information should be kept track of as the model is training.
# This value can be accessed through the logs parameter provided through the Keras callback functionality.
model.compile(optimizer = 'sgd', loss = 'mean_squared_error', metrics=['accuracy'])
```

----

### Training the Neural Network

```python
x = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# The Callback class is being used to stop training the model once it has reached a 95% accuracy. This can help prevent
# large waiting times due to the extra epochs that will run, but can also be used to prevent overfitting.
class Callback(tf.keras.callbacks.Callback):
    # The method must have the format: on_epoch_end(self, epoch, logs={})
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('accuracy') > 0.95:
            print('\n95% accuracy reached. Training will be terminated.')
            # The model attribute is used to communicate with the process.
            self.model.stop_training = True
callback = Callback()
            
# The instantiated callback object is passed in when fitting the training data to the model.
model.fit(x, y, epochs=500, callbacks = [callback])
```

----

### Predicting Values

```python
print(model.predict([10.0]))
```
